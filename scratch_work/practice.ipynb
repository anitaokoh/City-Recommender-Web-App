```python
import math
def cosim(x, y):
    
    """x and y are arrays / vectors"""
    
    num = 0
    for movie in great.columns:
        num += x[movie] * y[movie]
        
#     num = np.dot(x.values, y.values)
        
    xnorm = math.sqrt(sum(x ** 2))
    ynorm = math.sqrt(sum(y ** 2))
    
    denom = xnorm * ynorm
    
    return num/denom
```
